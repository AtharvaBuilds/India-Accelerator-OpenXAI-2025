```python
# Titanic Visualization Project

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")


df = pd.read_csv("/content/drive/MyDrive/titanic/train.csv")

print(df.head())


# Bar Plot: Survival count
plt.figure(figsize=(6,4))
sns.countplot(x="Survived", data=df, palette="viridis")
plt.title("Survival Count")
plt.show()

# Bar Plot: Survival by Gender
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", hue="Survived", data=df, palette="Set2")
plt.title("Survival by Gender")
plt.show()

# Box Plot: Age distribution by Survival
plt.figure(figsize=(6,4))
sns.boxplot(x="Survived", y="Age", data=df, palette="pastel")
plt.title("Age Distribution by Survival")
plt.show()

# KDE Plot: Age distribution
plt.figure(figsize=(6,4))
sns.kdeplot(df["Age"].dropna(), shade=True, color="blue")
plt.title("Age Distribution of Passengers")
plt.show()

# Violin Plot: Class vs Age
plt.figure(figsize=(8,5))
sns.violinplot(x="Pclass", y="Age", data=df, palette="muted")
plt.title("Age Distribution by Passenger Class")
plt.show()

# Swarm Plot: Fare by Class
plt.figure(figsize=(8,5))
sns.swarmplot(x="Pclass", y="Fare", data=df, palette="deep", size=3)
plt.title("Fare Distribution by Class")
plt.show()

#Summary
#1. Majority of passengers did not survive.
#2. Women had a much higher survival rate than men.
#3. Younger passengers (children) had better survival chances.
#4. First-class passengers were older on average and paid higher fares.
#5. Socio-economic status (class) and gender were strong factors in survival.

```
